## 6. Оценка качества моделей:

Метрики качества прогнозирования, такие как  R²(коэффициент детерминации),MAE (Mean Absolute Error), MSE (Mean Squared Error), RMSE (Root Mean Squared Error) и MAPE (Mean Absolute Percentage Error), играют важную роль в оценке 
эффективности моделей прогнозирования. Рассмотрим каждую из этих метрик:


1.  Обоснование использования метрики R² при прогнозировании цен криптовалют
Одной из ключевых метрик для оценки качества таких моделей является коэффициент детерминации — R².

Формула R²
R² определяется по следующей формуле:
  $$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$$
    
Преимущества коэффициента детерминации R² заключаются в том, что это нормированная метрика, принимающая значения от -∞ до 1. Это позволяет легко сравнивать различные модели между собой. Чем ближе значение R² к 1, тем лучше модель объясняет дисперсию данных, что делает его удобным для интерпретации. Кроме того, R² легко интегрируется в различные библиотеки машинного обучения, такие как scikit-learn.

Чтобы использовать R² для оценки модели, необходимо выполнить несколько шагов. Сначала нужно обучить модель на исторических данных, таких как цена и объем торгов. Затем следует сделать предсказания на тестовой выборке. После этого вычисляется значение R² между реальными и предсказанными значениями. Наконец, полученное значение R² позволяет оценить качество модели.
R² — удобная и информативная метрика, которая помогает понять, насколько хорошо модель прогнозирует цены на криптовалюты. Однако она должна использоваться вместе с другими метриками, такими как MAE или MSE, для более полной оценки.

2. MAE (Mean Absolute Error)

MAE измеряет среднее абсолютное отклонение прогнозируемых значений от фактических. Он рассчитывается по формуле:

$$ MAE =  \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y}_i| $$

где yᵢ — фактическое значение, ^yᵢ — прогнозируемое значение, а n — количество наблюдений.

MAE легко интерпретируется и дает представление о среднем размере ошибки в тех же единицах, что и данные. Это особенно важно для криптовалют, где небольшие изменения могут иметь значительное влияние на стоимость.

3. MSE (Mean Squared Error)

MSE измеряет среднее квадратов ошибок между фактическими и прогнозируемыми значениями:

$$ MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2 $$

MSE более чувствителен к большим ошибкам, так как ошибки возводятся в квадрат. Это может быть полезно в контексте криптовалют, где резкие колебания цен могут привести к значительным финансовым потерям. Меньшее значение MSE указывает на более высокую точность модели.

4. RMSE (Root Mean Squared Error)

RMSE является квадратным корнем из MSE:

$$ RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2} $$

RMSE также учитывает большие ошибки, но возвращает результат в тех же единицах, что и исходные данные. Это делает его более интерпретируемым по сравнению с MSE. В контексте криптовалют RMSE позволяет быстро оценить качество модели, учитывая влияние крупных колебаний цен.

5. MAPE (Mean Absolute Percentage Error)

MAPE измеряет среднюю абсолютную процентную ошибку:

$$ MAPE = \frac{100\%}{n}\sum_{i=1}^{n}\left|\frac{y_i - \hat{y}_i}{y_i}\right| $$

MAPE позволяет оценить точность модели в относительных величинах, что особенно полезно для криптовалют с высокой волатильностью. Эта метрика помогает понять, насколько хорошо модель предсказывает изменения в процентном соотношении, что может быть более информативным для инвесторов и трейдеров.

```python
from binance.client import Client
import pandas as pd

client = Client()

def get_crypto_data(symbol: str, start_date: str, end_date: str) -> pd.DataFrame:
    """Получение исторических данных криптовалюты с Binance"""
    klines = client.get_historical_klines(
        symbol=symbol,
        interval=Client.KLINE_INTERVAL_1DAY,
        start_str=start_date,
        end_str=end_date
    )
    
    cols = ["timestamp", "open", "high", "low", "close", "volume", 
             "close_time", "quote_vol", "trades", "taker_buy_vol", 
             "taker_quote_vol", "ignore"]
    
    data = pd.DataFrame(klines, columns=cols)
    data["date"] = pd.to_datetime(data["timestamp"], unit="ms")
    numeric_cols = ["open", "high", "low", "close", "volume"]
    data[numeric_cols] = data[numeric_cols].apply(pd.to_numeric, axis=1)
    
    return data[["date"] + numeric_cols]

# Получение данных
btc = get_crypto_data("BTCUSDT", "2020-03-17", "2025-03-17")
eth = get_crypto_data("ETHUSDT", "2020-03-17", "2025-03-17")
sol = get_crypto_data("SOLUSDT", "2020-03-17", "2025-03-17")

# Сохранение в CSV
btc.to_csv("bitcoin_binance_5y.csv", index=False)
eth.to_csv("ethereum_binance_5y.csv", index=False)
btc = pd.read_csv('bitcoin_binance_5y.csv')
eth = pd.read_csv('ethereum_binance_5y.csv')
```
